```python
import collections
import json
import itertools
import functools
from pathlib import Path

import numpy as np
import pandas as pd
```

```python
provincies = [
    "AB",
    "BC",
    "MB",
    "NB",
    "NL",
    "NS",
    "NT",
    "NU",
    "ON",
    "PE",
    "QC",
    "SK",
    "YT",
]
```

```python
raw_data_base_dir = Path('raw-data')
```

```python
station_set_by_prov = collections.defaultdict(set)
```

```python
for province in provincies:
    data_dir = raw_data_base_dir / province
    for path in data_dir.iterdir():
        with path.open() as flink:
            df = pd.read_csv(path, encoding='unicode_escape')

        station_set_by_prov[province].update(
            df['Climate ID']
        )
        del df
```

```python
for province, station_set in station_set_by_prov.items():
    print(province)
    display(station_set)
```

```python
all_station_ids = set()
all_station_ids.update(*station_set_by_prov.values())

display(all_station_ids)
```

```python
with Path('all-station-ids.json').open('w') as flink:
    json.dump(
        list(all_station_ids),
        flink
    )
```
