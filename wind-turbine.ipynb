```python
import multiprocessing

import pandas as pd

from helper import distance
```

```python
wind_turbine_df = pd.read_excel(
    'https://ftp.cartes.canada.ca/pub/nrcan_rncan/Wind-energy_Energie-eolienne/wind_turbines_database/Wind_Turbine_Database_FGP.xlsx'
)
wind_stations_df = pd.read_csv('https://dd.weather.gc.ca/climate/observations/climate_station_list.csv')
```

```python
def process_wind_turbin_distance_row(item_i):
    idx_i, row_i = item_i

    return [
        row_i['Turbine identifier'],
        *[
            0 if idx_i == idx_j else distance(
                lat1=row_i['Latitude'],
                lon1=row_i['Longitude'],
                lat2=row_j['Latitude'],
                lon2=row_j['Longitude'],
            )
            for idx_j, row_j in wind_turbine_df.iterrows()
        ]
    ]

with multiprocessing.Pool(processes=None) as pool:
    wind_turbine_distance_df = pd.DataFrame(
        columns=[
            'Turbine identifier',
            *[
                row_j['Turbine identifier']
                for _, row_j in wind_turbine_df.iterrows()
            ]
        ],
        data=pool.imap(
            func=process_wind_turbin_distance_row,
            iterable=wind_turbine_df.iterrows(),
            chunksize=16,
        )
    )
wind_turbine_distance_df.set_index('Turbine identifier', inplace=True)
```

```python
wind_turbine_distance_df.to_csv('wind_turbine_distance.csv')
```

```python
def process_wind_turbin_station_distance_row(item_i):
    idx_i, row_i = item_i

    return [
        row_i['Turbine identifier'],
        *[
            distance(
                lat1=row_i['Latitude'],
                lon1=row_i['Longitude'],
                lat2=row_j['Latitude'],
                lon2=row_j['Longitude'],
            )
            for idx_j, row_j in wind_stations_df.iterrows()
        ]
    ]

with multiprocessing.Pool(processes=None) as pool:
    wind_turbine_station_distance_df = pd.DataFrame(
        columns=[
            'Turbine identifier',
            *[
                row_j['Climate ID']
                for _, row_j in wind_stations_df.iterrows()
            ]
        ],
        data=pool.imap(
            func=process_wind_turbin_station_distance_row,
            iterable=wind_turbine_df.iterrows(),
            chunksize=16,
        )
    )
wind_turbine_station_distance_df.set_index('Turbine identifier', inplace=True)
```

```python
wind_turbine_station_distance_df.to_csv('wind_turbine_station_distance.csv')
```
