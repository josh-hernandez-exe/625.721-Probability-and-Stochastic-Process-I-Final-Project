```python
from collections import Counter
import json

import numpy as np
import pandas as pd
```

```python
wind_turbine_df = pd.read_excel(
    'https://ftp.cartes.canada.ca/pub/nrcan_rncan/Wind-energy_Energie-eolienne/wind_turbines_database/Wind_Turbine_Database_FGP.xlsx'
)
wind_stations_df = pd.read_csv('https://dd.weather.gc.ca/climate/observations/climate_station_list.csv')
```

```python
wind_stations_df['Year Duration'] = wind_stations_df['Last Year'] - wind_stations_df['First Year']
```

```python
currently_active = wind_stations_df['Last Year'] == 2023
```

```python
atleast_index = wind_stations_df['First Year'] <= 2010
```

```python
long_currently_active = currently_active * atleast_index
```

```python
longest_curently_active = wind_stations_df.loc[currently_active, 'Year Duration'].nlargest(n=20).index
```

```python
with open('all-station-ids.json') as flink:
    station_ids_with_data = json.load(flink)
    station_ids_with_data = [
        str(item)
        for item in station_ids_with_data
    ]
    station_ids_with_data_set = set(station_ids_with_data)
```

```python
# wind_stations_df.loc[longest_curently_active]
wind_stations_df.loc[currently_active].loc[atleast_index]
```

```python
wind_stations_df[wind_stations_df['Province'] == 'MANITOBA'].loc[currently_active].loc[atleast_index]
```

```python
wind_stations_distance_df = pd.read_csv('wind_station_distance.csv', index_col='Climate ID')
wind_turbine_distance_df = pd.read_csv('wind_turbine_distance.csv', index_col='Turbine identifier')
wind_turbine_station_distance_df = pd.read_csv('wind_turbine_station_distance.csv', index_col='Turbine identifier')
```

```python
np.fill_diagonal(wind_stations_distance_df.values, np.NaN)
np.fill_diagonal(wind_turbine_distance_df.values, np.NaN)
```

```python
station_id_columns = [
    item
    for item in wind_turbine_station_distance_df.columns
    if item in station_ids_with_data_set
]
```

```python
wind_stations_distance_df = wind_stations_distance_df.loc[
    wind_stations_distance_df.index.map(lambda item: item in station_ids_with_data_set),
    station_id_columns
]
```

```python
wind_turbine_station_distance_df = wind_turbine_station_distance_df.loc[:, station_id_columns]
```

```python
w_t_df = wind_turbine_distance_df.unstack()
w_t_df.index.rename(('src', 'dest'), inplace=True)
w_t_df.dropna(inplace=False)

w_s_df = wind_stations_distance_df.unstack()
w_s_df.index.rename(('src', 'dest'), inplace=True)
w_s_df.dropna(inplace=False)

w_t_s_df = wind_turbine_station_distance_df.unstack()
w_t_s_df.index.rename(('Climate ID', 'Turbine identifier'), inplace=True)
w_t_s_df.dropna(inplace=False)
```

```python
w_t_df.sort_values(inplace=True)
w_s_df.sort_values(inplace=True)
w_t_s_df.sort_values(inplace=True)
```

```python
num_rows = 500
station_mask = None

for (src_station_id, dst_station_id), _ in w_s_df.head(num_rows).items():
    temp_station = (
        (wind_stations_df['Climate ID'] == src_station_id)
        +
        (wind_stations_df['Climate ID'] == dst_station_id)
    )

    if station_mask is None:
        station_mask = temp_station
    else:
        station_mask += temp_station

with pd.option_context('display.max_rows', 2*num_rows, 'display.max_columns', 20):
    # display(w_s_df.head(num_rows))
    display(wind_stations_df[station_mask].loc[long_currently_active])
```

```python
num_rows = 1000
station_mask = None
turbine_mask = None

for (station_id, turbine_id), _ in w_t_s_df.head(num_rows).items():
    temp_station = wind_stations_df['Climate ID'] == station_id
    temp_turbine = wind_turbine_df['Turbine identifier'] == turbine_id

    if station_mask is None:
        station_mask = temp_station
    else:
        station_mask += temp_station

    if turbine_mask is None:
        turbine_mask = temp_turbine
    else:
        turbine_mask += temp_turbine

selected_stations = wind_stations_df[station_mask].loc[long_currently_active]

with pd.option_context('display.max_rows', len(selected_stations), 'display.max_columns', 20):
    # display(w_t_s_df.head(num_rows))
    display(selected_stations)
    # display(wind_turbine_df[turbine_mask])
```

```python
selected_station_to_turbine = wind_turbine_station_distance_df.loc[:,selected_stations['Climate ID']].idxmin(axis=0)
selected_station_set = set(selected_stations['Climate ID'].astype(str))
selected_turbine_set = set(selected_station_to_turbine.values)
display(selected_station_to_turbine)
```

```python
selected_stations_distnce_df = (
    wind_stations_distance_df.loc[
        wind_stations_distance_df.index.map(lambda item: item in selected_station_set),
        wind_stations_distance_df.columns.map(lambda item: item in selected_station_set),
    ]
)

display(selected_stations_distnce_df)
display(
    pd.concat(
        [
            selected_stations_distnce_df.idxmin(axis=0),
            selected_stations_distnce_df.min(axis=0),
        ],
        axis=1,
    )
)
```

```python
indexes = tuple(zip(
    selected_station_to_turbine.index,
    selected_station_to_turbine.values,
))

indexes = [
    idx for idx in indexes
    if idx in w_t_s_df.index
]

display(w_t_s_df.loc[indexes])
```

```python
selected_turbine_mask = wind_turbine_df['Turbine identifier'].apply(lambda item: item in selected_turbine_set)

display(wind_turbine_df[selected_turbine_mask])
```

```python
turbine_counter = Counter()
station_counter = Counter()

num_rows = 40

for (src_station_id, dst_station_id), _ in w_s_df.head(num_rows).items():
    station_counter[src_station_id] += 0.5
    station_counter[dst_station_id] += 0.5

for (src_turbine_id, dst_turbine_id), _ in w_t_df.head(num_rows).items():
    turbine_counter[src_turbine_id] += 0.5
    turbine_counter[dst_turbine_id] += 0.5

for (station_id, turbine_id), _ in w_t_s_df.head(num_rows).items():
    station_counter[station_id] += 2
    turbine_counter[turbine_id] += 1


turbine_counter_df = pd.DataFrame(data=turbine_counter.items()).sort_values(by=1, ignore_index=True)
station_counter_df = pd.DataFrame(data=station_counter.items()).sort_values(by=1, ignore_index=True)

display(turbine_counter_df.tail(n=10))
display(station_counter_df.tail(n=10))
```
